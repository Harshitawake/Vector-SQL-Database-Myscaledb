```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
appDescription_embeddings=model.encode(df['appDescription'].to_list())
```

```python
df['appPlaceBetInfo'].isna().sum()
```

```python
df['appPlaceBetInfo'].fillna('',inplace=True)
```

```python
appPlaceBetInfo_embeddings=model.encode(df['appPlaceBetInfo'].to_list())
```

```python
type(appPlaceBetInfo_embeddings)
```

```python
appDescription_embeddings.shape
```

```python
import torch
torch.cuda.is_available()
device='cuda'
```

```python
# Convert each embedding to a list
appDescription_embeddings_list = [list(embedding) for embedding in appDescription_embeddings]
appPlaceBetInfo_embeddings_list = [list(embedding) for embedding in appPlaceBetInfo_embeddings]
```

```python

df1 = pd.DataFrame({

    'appDescription_embedding': appDescription_embeddings_list,
    'appPlaceBetInfo_embedding': appPlaceBetInfo_embeddings_list
})
```
